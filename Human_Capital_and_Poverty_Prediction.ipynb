```python
#splitting the data into X and Y
x=df[['Human Capital Investement (HCI)','Life expectancy at birth, total (years)',
       'School enrollment, primary (gross), gender parity index (GPI)']]
y=df["Real consumption of households and government"]
```

```python
#scaling the data using Z-score
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_data=sc.fit_transform(x)
scaled_data
```

```python
#using train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(scaled_data,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
X_test.shape
```

# **Developing the Model**

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Initialize the models
linear_model = LinearRegression()
ridge_model = Ridge()
rf_model = RandomForestRegressor()
```

```python
# Train the models
linear_model.fit(X_train, y_train)
ridge_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_linear = linear_model.predict(X_test)
y_pred_ridge = ridge_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)
```

```python
# Evaluate the models
def evaluate_model(y_test, y_pred, model_name):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{model_name} Performance:")
    print(f"Mean Squared Error: {mse:.2f}")
    print(f"R^2 Score: {r2:.2f}")
    print("")

evaluate_model(y_test, y_pred_linear, "Linear Regression")
evaluate_model(y_test, y_pred_ridge, "Ridge Regression")
evaluate_model(y_test, y_pred_rf, "Random Forest Regressor")
```
